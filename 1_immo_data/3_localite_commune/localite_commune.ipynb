```python
import pandas as pd
import re

loca_comm = pd.read_csv('Raw_localite_commune.csv', sep=';')
```

```python
def clean_loca_comm(df):
    df = df.drop(columns=['Zusatzziffer', 'BFS-Nr', 'E', 'N'])
    df = df.rename(columns={'Ortschaftsname': 'localite', 'PLZ': 'code postal', 'Gemeindename': 'commune', 'Kantonskürzel': 'canton', 'Sprache': 'langue'})


    return df
```

```python
localite_commune = clean_loca_comm(loca_comm)
localite_commune.head(10)
```

```python
#Création du nouveau fichier CSV nettoyé
localite_commune.to_csv('Clean_localite_commune.csv', index=False)
```
